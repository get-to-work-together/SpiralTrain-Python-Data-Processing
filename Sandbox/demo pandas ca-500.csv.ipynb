# Demo Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

### Read the data

```python
filename = '../datasets/ca-500.csv'
df = pd.read_csv(filename)
df
```

### Exploratory Data Analysis

```python
df.info()
```

```python
df.describe()
```

### Convert data types

```python
df = df.astype('string')

df['city'] = df['city'].astype('category')
df['province'] = df['province'].astype('category')

df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], yearfirst = True)

df.dtypes
```

## Hierarchical index

```python
df.set_index(['province', 'city']).sort_index()
```

## Remove Missing data

```python
df.count()
```

```python
missing = df[df['email'].isna()]

print(f'{missing.shape[0]} rows with no email')
missing
```

#### identical to

```python
df[np.any(df.isna(), axis=1)]
```

```python
df.dropna(subset=['email'], inplace = True)
```

```python
df.count()
```

## Remove Duplicates

```python
df[df.duplicated()]
```

```python
duplicates = df[df.duplicated(['last_name', 'first_name','email'])]
duplicates
```

```python
df[df['email'].isin(duplicates['email'])].sort_values('last_name')
```

```python
df.drop_duplicates(['last_name', 'first_name', 'email'], keep = 'last', inplace=True)
```

```python
df.count()
```

### Aggregate Cities

```python
len(df['city'].unique())
```

```python
cities = sorted(df['city'].unique())

print('%d cities' % len(cities))
for city in cities:
    print(city)
```

```python
df['city'].value_counts().head(10)
```

### Aggregate Provinces

```python
df_provinces = pd.DataFrame(df['province'].value_counts())
df_provinces
```

```python
df['province'].value_counts(ascending=True).plot(kind='bar')
```

### Join with Province Names

```python
d = {'BC': 'British Columbia',
     'AB': 'Alberta',
     'SK': 'Saskatchewan',
     'MB': 'Manitoba',
     'ON': 'Ontario',
     'QC': 'Qubec',
     'NB': 'New Brunswick',
     'PE': 'Prince Edward Island',
     'NS': 'Nova Scotia',
     'NL': 'Newfoundland & Labrador'}

provinces_names = pd.DataFrame({'ProvinceName': d}).sort_index()
provinces_names
```

```python
df_provinces = df_provinces.join(provinces_names)
df_provinces
```

```python
df_provinces.set_index('ProvinceName').plot(kind='barh')
plt.show()
```
