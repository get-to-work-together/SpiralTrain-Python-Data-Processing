```python
import numpy as np
np.random.seed(19680801)
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```

### Box Plot

```python
plt.boxplot(x)
plt.show()
```

```python
median = np.median(x)
q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = x[(x < lower) | (x > upper)]
minimum = np.min(x[(x >= lower) & (x <= upper)])
maximum = np.max(x[(x >= lower) & (x <= upper)])

d = {'median': median,
     'q1': q1,
     'q3': q3,
     'iqr': iqr,
     'lower': lower,
     'upper': upper,
     'number of outliers': len(outliers),
     'minimum': minimum,
     'maximum': maximum}

for k, v in d.items():
    if isinstance(v, int):
        print(f'{k:20} : {v:4}')
    else:
        print(f'{k:20} : {v:8.3f}')

print( 'outliers', np.sort(outliers).round(2) )
```

### Named Colors

```python
"""
========================
Visualizing named colors
========================

Simple plot example with the named colors and its visual representation.
"""
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

### Subplots

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(3*x)
y4 = np.sin(4*x)
```

```python
# MATLAB style interface

plt.subplot(2,2,1)
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.plot(x, y3)

plt.subplot(2,2,4)
plt.plot(x, y4)

plt.show()
```

```python
# Object oriented interface

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2, label = 'B')
ax[1, 0].plot(x, y3, label = 'C')
ax[1, 1].plot(x, y4, label = 'D')

plt.show()
```

```python
# Trystan interface

fig = plt.figure(figsize = (12, 8))

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y1, label = 'A')

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y2, label = 'B')

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, y3, label = 'C')

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, y4, label = 'D')

plt.legend()

plt.show()
```

### Seaborn

```python
import seaborn as sns
sns.set(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", 
            y="tip",
            hue="smoker", 
            palette=["m", "g"],
            data=tips)

sns.despine(offset=10, trim=True)
```
